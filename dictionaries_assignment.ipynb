
### Assignment 14: Counting with Dictionaries

Write a function that takes a string and returns a dictionary with the count of each character in the string. Print the dictionary.


```python
def char_count(input_string):
    """
    Creates a dictionary with the frequency of each char in the string
    """
    dictionary = {}     #{H,e,l}
    for char in input_string:
        if char in dictionary.keys():
            dictionary[char] += 1
        else:
            dictionary[char]= 1
    return dictionary

print(char_count("Hello"))

```

```python
#binary search with recursion

def binary_recursion_helper(arr,target,s,e):
    #edge cases
    if s>e:
        return False
    m = (s+e)//2
    if arr[m]==target:
        return True
    if arr[m]<target:
        s=m+1
    else:
        e=m-1
    ans = binary_recursion_helper(arr,target,s,e)
    if ans==True:
        return True
    if ans==False:
        return False

def binary_recursion(arr,target):
    return binary_recursion_helper(arr,target,s=0,e=(len(arr)-1))

print(binary_recursion([3,4,6,7,8,10],8))
```

```python
#Merge Sort using Recursion

#Merges first and second half of list by comparing value
#Adds to Ans
#Overwrites L1 values with Ans values
def Merge(l1,s,m,e):
    i = s
    j = m+1
    ans = []
    while i<=m and j<=e:
        if l1[i]<l1[j]:
            ans.append(l1[i])
            i+=1
        elif l1[i]>l1[j]:
            ans.append(l1[j])
            j+=1
        elif l1[i]==l1[j]:
            ans.append(l1[i])
            ans.append(l1[j])
            i+=1
            j+=1
    while i<=m:
        ans.append(l1[i])
        i+=1
    while j<=e:
        ans.append(l1[j])
        j+=1
        
    startOfMyAns = 0
    startOfMyList = s
    
    while startOfMyList <= e:
        l1[startOfMyList] = ans[startOfMyAns]
        startOfMyAns+=1
        startOfMyList+=1    
    return

#Splits list in 2
#Ends with list of 2 items
#Then calls merge function to merge the 2 items
def MergeSortHelper(l1,s,e):
    if s>=e:
        return
    m = (s+e)//2
    MergeSortHelper(l1,s,m)
    MergeSortHelper(l1,m+1,e)
    Merge(l1,s,m,e)
    return

def MergeSort(l1):
    return MergeSortHelper(l1,0,len(l1)-1)
    

l1 = [6,5,12,10,9,1]
MergeSort(l1)
print(l1)
```

```python
#Quick Sort
"""
finding rp of pivot item.
if l1[i]>pivot? do nothing, rp would be 0
if l1[i]<pivot - move rp += 1

now l1[rp] = pivot but need to swap to keep other number.
then can repear process with new partition #

"""

def PartitionFunction(l1,s,e):
    #designate last list item as pivot (although could be any item in list)
    pivot = l1[e]
    #initialize rightPosition, and initialize i as index in list which function will loop through until end  
    rightPosition = 0
    i = 0
    #find right position for pivot number
    while i<len(l1):
        if l1[i]<pivot:
            rightPosition += 1
        i+=1
    #swap pivot number with number in its correct position
    l1[e], l1[rightPosition] = l1[rightPosition], l1[e]

    pivotIndex = rightPosition

    #work we want to do is if we have an item greater than pivot from first sort and one less than pivot from second sort, swap them.
    while (s<pivotIndex and e>pivotIndex):
        if l1[s] <= l1[pivotIndex]:
            s+=1
        elif l1[s] > l1[pivotIndex]:
            if l1[e] < l1[pivotIndex]:
                l1[s],l1[e] = l1[e],l1[s]
                s+=1
                e-=1
            elif l1[e] >= l1[pivotIndex]:
                e-=1
    return pivotIndex

def QuickSort(l1,s,e):
    if(s>=e):
        return

    pivotIndex = PartitionFunction(l1,s,e)
    #split list into area before pivot and area after pivot
    QuickSort(l1,s,pivotIndex-1)
    QuickSort(l1,pivotIndex+1,e)
    return

l1 = [3,6,7,2,1,4,5]

QuickSort(l1,0,len(l1)-1)
print(l1)
```

```python
phone_dict = {'1':'','2':'abc','3':'def','4':'ghi','5':'jkl','6':'mno','7':'pqrs','8':'tuv','9':'wxyz'}

def phone_combinations(s):
    if s=="":
        return [""]
    currentChar = s[0]
    smallOptions = phone_combinations(s[1:]) #assume funct will add next option each possible combination option; and narrow down to last digit
    #my work
    optionsList = []

    for itemChar in phone_dict[currentChar]: #like a in a,b,c from dict
        #add this option to all options in smallOptions
        for option in smallOptions:
            optionsList.append(itemChar + option)

    return optionsList

ans = phone_combinations("23")
print(ans)
```

```python
# print linked list
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

first = Node(1)
second = Node(2)
third = Node(3)

first.next = second
second.next = third


def print_ll(first):
    head = first
    while head != None:
        print(head.data)
        head = head.next

print_ll(first)
```

```python
# print linked list through recursion
def print_ll_recursive(head):
#base case: head = None
    if head==None:
        return
#work
    print(head.data)
#assume recursion will print list items
    return print_ll_recursive(head.next)

print_ll_recursive(first)
```
